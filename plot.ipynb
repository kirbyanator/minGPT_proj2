```python
import matplotlib.pyplot as plt
```

```python
with open('losses.txt', 'w') as output_file:
    with open("output_results.txt", "r") as file:
        for line in file:
            if "loss" in line:
                item = line[-8:]
                output_file.write(item)
```

```python
losses = []
with open("losses.txt", "r") as loss_file:
    for line in loss_file:
        losses.append(float(line))
print(losses)
```

```python
plt.title("Losses over 300k iterations")
plt.xlabel("K Iterations")
plt.ylabel("Cross-Entropy Loss")
plt.plot(losses)
plt.savefig("losses.jpg", bbox_inches='tight')
```

```python
normal_losses = []
ul2_losses = []

with open("normal_losses.txt") as file:
    for line in file:
        normal_losses.append(float(line))

with open("ul2_losses.txt") as file:
    for line in file:
        ul2_losses.append(float(line))
print(normal_losses)
print(ul2_losses)
```

```python
plt.title("Losses of normal model vs ul2 model over 500 iterations")
plt.xlabel("iterations")
plt.ylabel("losses")
plt.plot(normal_losses, label="normal losses")
plt.plot(ul2_losses, label="ul2 losses")
plt.legend()
plt.savefig("loss_comparison.jpg", bbox_inches='tight')
```

```python
normal_score = .2579
normal_iters = 500
ul2_score = .2590
ul2_iters = 50

plt.title("performance of model on 'hellaswag' task")
plt.xlabel("iterations")
plt.ylabel("accuracy")
plt.plot((normal_iters, normal_score))
plt.plot((ul2_iters, ul2_score))
plt.show()
```
